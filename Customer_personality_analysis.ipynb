```python
df_copy=df.copy()
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
df_copy.iloc[:, 2:] = sc.fit_transform(df_copy.iloc[:, 2:])
df_copy
```

```python
plt.rcParams["figure.figsize"] = [20, 10]
ax = sns.heatmap(df_copy.corr(),cmap = 'coolwarm', annot=True,annot_kws = {'size': 8},vmax=1,vmin=-1,fmt='0.2f', linewidths=0.25)
#plt.savefig("heatmap")
plt.show()
```

We can see most of the features are highly correlated

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 0.95)
pca.fit(df_copy)
reduced_data = pd.DataFrame(pca.transform(df_copy))

print("Number of features after PCA= ",len(reduced_data.columns))
```

```python
reduced_data
```

```python
#calculate distortions for different values of k (number of clusters)
from sklearn.cluster import KMeans
distortions = []
K = range(1,15)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(reduced_data)
    distortions.append(kmeanModel.inertia_)
```

```python
#plot elbow graph
plt.figure(figsize=(20,10))
plt.grid()
plt.plot(K, distortions, 'go-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method to find optimal K')

plt.show()
```

from elbow plot , K can be taken as 2

```python
clusters = KMeans(n_clusters=2, random_state=1).fit(reduced_data)
df["Cluster"]= clusters.labels_
df.head()
```

```python
plt.figure(figsize=(7,5))

df['Cluster'].value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.legend(["Cluster_1","Cluster_0"],bbox_to_anchor=(1.02, 1),title="Clusters")

plt.show()
```

```python
sns.set_theme(style="whitegrid")
ax=sns.countplot(x=df["Cluster"])
for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.35, p.get_height()-70))
plt.show()
```

```python
plt.figure(figsize = (15,7) )
sns.kdeplot(data =df, x = 'Age' , hue = 'Cluster')
plt.title('Age Distribution' ,fontsize = 15)
plt.show()
```

```python
clus0 = df[(df['Age'] > 50 )&(df['Cluster'] == 0)]
clus1 = df[(df['Age'] > 50 )&(df['Cluster'] == 1)]
print('percentage of customer more than 50 at cluster0 : {0:0.2f}%'.format(len(clus0) / len(df[df['Cluster'] == 0])))
print('percentage of customer more than 50 at cluster1 : {0:0.2f}%'.format(len(clus1) / len(df[df['Cluster'] == 1])))
```

```python
plt.figure(figsize = (20,10) )
labels = ['Fruit' , 'Meat' , 'Fish' , 'Sweets' , 'Wine' , 'Gold']
for i in range(2):
    counts =[]
    d = df[df['Cluster'] == i]
    counts.append(d['MntFruits'].sum())
    counts.append(d['MntMeatProducts'].sum())    
    counts.append(d['MntFishProducts'].sum())    
    counts.append(d['MntSweetProducts'].sum())    
    counts.append(d['MntWines'].sum())
    counts.append(d['MntGoldProds'].sum())
    plt.subplot(1,2,i+1)
    plt.title("Cluster "+str(i))
    plt.pie(counts, labels = labels, autopct='%.0f%%' )
plt.show()
```
